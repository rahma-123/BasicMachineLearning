```python
#Menggunakan data sebagai input
#menghitung total belanja sebelum diskon

import pandas as pd

data = { 'Barang' : ['Laptop' , 'Mouse' , 'Keyboard' , 'Headset', 'Flashdisk'] , 'total_belanja ' : [7000000, 150000, 300000, 250000, 90000 ] }
df = pd.DataFrame(data)
df
```

```python
#menghitung total belanja
total_belanja = df['total_belanja '].sum()
print(f"Total Belanja Sebelum Diskon : Rp{total_belanja:,.0f}")  # Format dengan pemisah ribuan
```

```python
# Fungsi untuk menghitung diskon berdasarkan total belanja
def hitung_diskon_berdasarkan_total(total_belanja):



    # Menentukan persen diskon berdasarkan total belanja
    if total_belanja >= 1000000:
        persen_diskon = 25
    elif total_belanja >= 500000:
        persen_diskon = 15
    elif total_belanja >= 250000:
        persen_diskon = 10
    elif total_belanja >= 100000:
        persen_diskon = 5
    else:
        persen_diskon = 0

    # Menghitung jumlah diskon
    diskon = total_belanja * (persen_diskon / 100)




    # Menghitung harga setelah diskon
    harga_setelah_diskon = total_belanja - diskon

    # Menampilkan hasil
    print("\n================================")
    print(f"Total Belanja : Rp{total_belanja:,.0f}")  # Format dengan pemisah ribuan
    print(f"Diskon : [{persen_diskon}%]")
    print(f"Harga Setelah Diskon : Rp{harga_setelah_diskon:,.0f}")  # Format dengan pemisah ribuan

# Contoh pemanggilan fungsi dengan beberapa nilai total belanja
hitung_diskon_berdasarkan_total(7000000)
hitung_diskon_berdasarkan_total(150000)
hitung_diskon_berdasarkan_total(300000)
hitung_diskon_berdasarkan_total(250000)
hitung_diskon_berdasarkan_total(90000)
hitung_diskon_berdasarkan_total(7790000)
```

## **SOAL LATIHAN 7**

```python
#menggunakan dataset yang telah disediakan

data_siswa = {
    'Jam_Belajar' : [2,3,4,5,6,7,8,9,10,12] ,
    'Nilai_Akhir' :[50, 55, 60, 65, 70, 75, 78, 80, 85, 90]
}

df = pd.DataFrame(data_siswa)
df


```

```python
#menggunakan scikit-learn untuk membuat regresi linier

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# mmisahkan fitur (Jam Belajar) dan target (Nilai Akhir)
x = df[['Jam_Belajar']]  # Fitur (independen)
y = df['Nilai_Akhir']    # Target (dependen)

# Membagi data menjadi data latih dan data uji (80% latih, 20% uji)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)



#membuat model regresi linier
model = LinearRegression()

#melatih model dengan data latih
model.fit(x_train, y_train)

# mengevaluasi model (koefisien dan intercept)
print(f'Koefisien (Slope): {model.coef_}')
print(f'Intercept (Intercept): {model.intercept_}')


# Prediksi nilai menggunakan data uji
y_pred = model.predict(x_test)
# Menampilkan hasil prediksi dan nilai sebenarnya
print(f'Nilai Sebenarnya (y_test): {y_test.values}')
print(f'Prediksi Nilai (y_pred): {y_pred}')
```

```python
#membuat scatter plot untuk visualisasi hasil

# Menampilkan plot untuk visualisasi model regresi
plt.scatter(x, y, color='Blue', label='Data Siswa')  # Data asli
plt.plot(x, model.predict(x), color='red', label='Model Regresi Linier')  # Model regresi linier
plt.xlabel('Jam Belajar')
plt.ylabel('Nilai Akhir')
plt.title('Regresi Linier: Jam Belajar vs Nilai Akhir')
plt.legend()
plt.show()
```

```python
#mencari hubungan X dan Y

X = df ['Jam_Belajar']
Slope = model.coef_[0]
y = model.intercept_ + (Slope * X)
df = pd.DataFrame({'Jam_Belajar': X, 'Nilai_Akhir': y})
df

print(f"Persamaan Regresi : y = {Slope:}X + {model.intercept_ :}")
```

# **# SOAL LATIHAN 8**

```python
#membuat dafa frame

import pandas as pd

data_rumah = {
    'Luas_Tanah' : [100,200,150,250,300] ,
    'Jumlah_Kamar' : [2,3,3,4,5] ,
    'Lokasi' : ['A' , 'B' , 'A' , 'C' , 'B'] ,
    'Harga' : [500,700,600,850,900]
 }

df = pd.DataFrame(data_rumah)
df
```

```python
#menghitung ra-rata dari harga rumah

import numpy as np

rata_rata_harga = np.mean(df['Harga'])
print ("rata_rata_harga : " , np.mean(df['Harga']))
```

```python
#membuat scatter plot antara luas tanah dan harga rumah menggunakan matplotlib

import matplotlib.pyplot as plt

plt.scatter(df['Luas_Tanah'] , df['Harga'])
plt.xlabel('Luas_Tanah')
plt.ylabel('Harga')
plt.title('Grafik Luas Tanah dan Harga Rumah')
plt.show()
```

```python
#membuat Scikit Learn Untuk model regresi Linier untuk presiksi harga rumah

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# mmisahkan fitur (Luas Tanah) dan target (Harga)
x = df[['Luas_Tanah']]  # Fitur (independen)
y = df['Harga']         # Target (dependen)



#membagi data menjadi data latih dan uji (80% Latih, 20% Uji)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,random_state=42 )

#membuat model regresi linear
model = LinearRegression()

#melatih model dengan data latih
model.fit(x_train, y_train)


# mengevaluasi model (koefisien dan intercept)
print(f'Koefisien (Slope): {model.coef_}')
print(f'Intercept (Intercept): {model.intercept_}')


# Prediksi nilai menggunakan data uji
y_pred = model.predict(x_test)
# Menampilkan hasil prediksi dan nilai sebenarnya
print(f'Nilai Sebenarnya (y_test): {y_test.values}')
print(f'Prediksi Nilai (y_pred): {y_pred}')

```

```python
#mencari hubungan X dan Y

X = df ['Luas_Tanah']
Slope = model.coef_[0]
y = model.intercept_ + (Slope * X)
df = pd.DataFrame({'Luas_Tanah': X, 'Harga': y})
df

print(f"Persamaan Regresi : y = {Slope:}X + {model.intercept_ :}")
```

```python
#menggunakan model untuk memprediksi harga rumah dengan luas tanah 180m^2


Luas_Tanah_Baru = 180
prediksi_harga = model.intercept_ + (Slope * Luas_Tanah_Baru)

print(f"Prediksi Harga Rumah dengan Luas Tanah {Luas_Tanah_Baru}m^2 adalah : Rp{prediksi_harga}")
```
